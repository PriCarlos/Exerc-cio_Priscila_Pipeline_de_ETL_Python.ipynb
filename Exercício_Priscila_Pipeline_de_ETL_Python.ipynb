# **Extract (Extração)**
Extraindo os IDs do arquivo SCV. Com o Pandas.

```python
import pandas as pd
tabela = pd.read_csv("/cancelamentos.csv")
display(tabela)
```

# **Transform (Transformação):**

Achei desnecessária a coluna "CustomerID" para essa análise, e dei um drop() nela. Também removi os valores vazios da minha tabela com dropna() e verifiquei a remoção com isna().sum()

```python
tabela = tabela.drop("CustomerID", axis = 1)
display(tabela)
```

```python
display(tabela.info())
tabela =  tabela.dropna()
display(tabela.info())
```

```python
#print(tabela.isna().sum())
```

**Análise inicial**

Analisei as tabelas “cancelou” e “duracao_contrato”, para entender em qual cenário aconteciam os cancelamentos. Também dei um “groupby” em “assinaturas” e “duracao_contrato” para tirar a média entre as duas colunas. Percebi que os contratos mensais sempre cancelam. Removi, assim os contratos mensais da coluna “duracao_contrato”  e refiz a análise anterior e constatei que, acabando com os planos mensais, o número de cancelamentos diminuiu e o número de não cancelamento aumentou.


```python
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.1%}".format))
```

```python
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
display(tabela.groupby("assinatura").mean(numeric_only=True))
```

```python
tabela = tabela[tabela["duracao_contrato"]!="Monthly"]
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.1%}".format))
```
